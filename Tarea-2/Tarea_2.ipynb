A continuación, se presentan las 4 categorías en las que se dividen los datos: ['Tipo de supermercado1', 'Tipo de supermercado2', 'Tienda de comestibles', 'Tipo de supermercado3'].

##Funciones de visualizacion

*   **distribution_plot:** se crea función de usuario para graficar tanto las medidas estadísticas como el histograma de una variable, esto para poder reutilizar ya que se va necesitar en varias sesiones del documento usa la libreria **plotly.**



```python
def distribution_plot(dataframe, variable, titulo):
    fig = px.histogram(dataframe,
                       x=variable,
                       marginal='box',
                       title='Distribución de '+ variable + titulo)
    fig.update_layout(bargap=0.1)
    fig.show()
```

# Manipulación dataset

##Eliminar filas con nulos
Uno de los métodos para preprocesado de dataset es eliminar las filas o columnas que contengan valores nulos para este caso primero vamos a ver cuántos valores nulos existe en el dataset por cada columna.
Por medio de **isna()** y **sum()** obtener el detalle de nuestro dataset.

```python
df.isna().sum()
```

### Aplicando la eliminación
**Paso 1:** Copiamos el df para no cambiar el dataset original por medio de **copy()**

**Paso 2:** Por medio de **dropna()** podemos eliminar toda fila por el parametro **(axis = 0)** que contenga algún campo en nulo en sus columnas.

**Paso 3:** Mediante las funciones **describe y isna** revisamos que nos quedó del dataset.

**Importante este método es el menos recomendado ya que como observamos elimino el 68% de los datos**

```python
df_droprow = df.copy()
df_droprow = df_droprow.dropna (axis = 0)
print(df_droprow.isna().sum())
df_droprow.describe()
```

```python
distribution_plot(df, 'Item_Weight',' antes del reemplazo')
```

##Reemplazando con la media
Consiste en reemplazar los valores numéricos faltantes con la media de cada columna o bien si se conoce bastante los datos se puede granular y calcular la media según una o mas variables categóricas para ser más precisos. Para esta parte vamos a reemplazar con la media general de la columna con los valores faltantes.

### Aplicando el reemplazo
**Paso 1:** Para este caso se usará también la librería **numpy** que se complementa bastante con **pandas**.

**Paso 2:** Igual que en el caso de eliminación de filas se crea una copia del dataset.

**Paso 3:** Creamos un dataset con las columnas numéricas ya que las categóricas no se les puede aplicar el proceso de media y luego hacemos un dataset con el nombre de esas columnas.

**Paso 4:** Se indica al dataset que se va a usar solo las columnas numéricas ingresado la variable `numeric_columns` y por medio de la función **fillna** reemplazamos todo valor "nan" por la media correspondiente.

**Paso 5:** Por último vemos el resultado de los cambios como hemos venido haciendo por medio de **describe y isna.**

```python
df_mean = df.copy()
numeric = df_mean.select_dtypes(include=np.number)
numeric_columns = numeric.columns
df_mean[numeric_columns] = df_mean[numeric_columns].fillna(df_mean.mean())

print(df_mean.isna().sum())
df_mean.describe()

#Dataframe a usar en el futuro
df_pre = df.copy()
numeric = df_mean.select_dtypes(include=np.number)
numeric_columns = numeric.columns
df_pre[numeric_columns] = df_pre[numeric_columns].fillna(df_pre.mean())
```

```python
distribution_plot(df_mean, 'Item_Weight' , 'despues del reemplazo')
```

##Reemplazando con la mediana
Al igual que el proceso de **media** aplicamos cada uno de los pasos con la diferencia que no usamos la función de media `mean` sino que usamos la función **median()**

```python
df_median = df.copy()
numeric = df_median.select_dtypes(include=np.number)
numeric_columns = numeric.columns
df_median[numeric_columns] = df_median[numeric_columns].fillna(df_median.median())

print(df_median.isna().sum())
df_median.describe()
```

```python
distribution_plot(df_median, 'Item_Weight' , 'despues del reemplazo')
```

##Agrupar valores categóricos con mismo significado
Como notamos la columna **'Item_Fat_Content'** define 'Low Fat' con dos formas de escritura entonces unificamos reemplazando el LF que es el acrónimo, mismo caso para 'Regular'.

**Paso 1:** Contar los valores por categoría para demostrar el problema en el dataset por medio de la función **value_counts()**.

**Paso 2:** Aplicar el reemplazo en la columna con la función **replace()**.

**Paso 3:** Verificar que haya quedado el cambio aplicado según lo esperado.

```python
df_agrupar = df.copy()
print("ANTES DE UNIFICAR:\n",df_agrupar['Item_Fat_Content'].value_counts())
sns.catplot(x='Item_Fat_Content', kind='count', data=df_agrupar)

df_agrupar.replace({'Item_Fat_Content':{'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'}},inplace=True)
print("\n\nDESPUES DE UNIFICAR:\n",df_agrupar['Item_Fat_Content'].value_counts())
sns.catplot(x='Item_Fat_Content', kind='count', data=df_agrupar)

df_pre.replace({'Item_Fat_Content':{'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'}},inplace=True)

#Reemplazando valores de 'Outlet Size' con la moda
df_pre['Outlet_Size'] = df_pre['Outlet_Size'].fillna(df_pre['Outlet_Size'].mode()[0])
print("\n\nDESPUES DE UNIFICAR:\n",df_pre['Outlet_Size'].value_counts())
sns.catplot(x='Outlet_Size', kind='count', data=df_pre)
```

##Encodear datos categóricos a valores numéricos

En seguida se encodearán los valores de las variables categóricas en valores numéricos.




```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_pre['Item_Fat_Content'] = encoder.fit_transform(df_pre['Item_Fat_Content'])
df_pre['Item_Type'] = encoder.fit_transform(df_pre['Item_Type'])
df_pre['Outlet_Size'] = encoder.fit_transform(df_pre['Outlet_Size'])
df_pre['Outlet_Location_Type'] = encoder.fit_transform(df_pre['Outlet_Location_Type'])
df_pre['Outlet_Type'] = encoder.fit_transform(df_pre['Outlet_Type'])

df_pre.head()
```
